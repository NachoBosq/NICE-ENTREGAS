8. Crea un gr치fico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sof칤a mide 190 cm, Carmen mide 170 cm. Las barras ser치n verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es m치s sencillo dejarlos dentro de la figura.

```python
nombres = ["Ana", "Luis", "Pedro", "Sofia", "Carmen"]
alturas = [160, 180, 175, 190, 170]

fig, ax = plt.subplots()
ax.bar(x = nombres, height = alturas, color= "g")
ax.set_xlabel("Personas")
ax.set_ylabel("Alturas")
ax.set_title("Alturas grupo amigos")
```

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
nombres = ["Ana", "Luis", "Pedro", "Sofia", "Carmen"]
alturas = [160, 180, 175, 190, 170]

fig, ax = plt.subplots()

barplot = ax.bar(x=nombres, height=alturas, color="g")

for bar in barplot:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom')

ax.set_xlabel("Personas")
ax.set_ylabel("Alturas")
ax.set_title("Alturas grupo amigos")

plt.show()
```

12. Prueba a hacer el gr치fico con las barras en horizontal (no es necesario poner el texto al final de la barra).

```python
nombres = ["Ana", "Luis", "Pedro", "Sofia", "Carmen"]
alturas = [160, 180, 175, 190, 170]

fig, ax = plt.subplots()

barplot = ax.barh(y=nombres, width=alturas, color="b")

for bar in barplot:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height() / 2,str(width),va='center')

ax.set_xlabel("Alturas")
ax.set_ylabel("Personas")
ax.set_title("Alturas grupo amigos")
ax.set_xticklabels([])

plt.show()
```

13. Prueba a invertir los ejes con ax.invert_axis()

```python
nombres = ["Ana", "Luis", "Pedro", "Sofia", "Carmen"]
alturas = [160, 180, 175, 190, 170]

fig, ax = plt.subplots()

barplot = ax.barh(y=nombres, width=alturas, color="b")

for bar in barplot:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height() / 2,str(width),va='center')

ax.invert_yaxis()

plt.show()
```
