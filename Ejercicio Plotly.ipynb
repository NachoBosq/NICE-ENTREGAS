## 1. Create an interactive bar chart showing total sales by State for2014.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df2014 = df[df['Order Date'].dt.year == 2014]

ventas_por_estado_2014 = df2014.groupby('State')['Sales'].sum().reset_index()
ventas_por_estado_2014 = ventas_por_estado_2014.sort_values(by='Sales', ascending=False)

trace1 = go.Bar(
    x=ventas_por_estado_2014['State'],
    y=ventas_por_estado_2014['Sales'],
    name='Total sales in 2014',
    marker=dict(color='rgba(0, 0, 255, 0.5)'),
    textangle=-45,  # Ajusta el ángulo de los textos de los estados en diagonal
    
)

data = [trace1]

layout = go.Layout(barmode="group")

fig = go.Figure(data=data, layout=layout)
fig.show()
```

## 2. Create an interactive line chart showing Sales and Profit in 2014.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df_2014 = df[df['Order Date'].dt.year == 2014]

trace_sales = go.Scatter(
    x=df_2014['Order Date'],
    y=df_2014['Sales'],
    name='Sales',
    mode='lines',
    line=dict(color='blue')
)

trace_profit = go.Scatter(
    x=df_2014['Order Date'],
    y=df_2014['Profit'],
    name='Profit',
    mode='lines',
    line=dict(color='red')
)

# Define layout
layout = go.Layout(
    title='Sales and Profit in 2014',
    xaxis=dict(title='Order Date'),
    yaxis=dict(title='Amount'),
    hovermode="closest"

)

# Create figure
fig = go.Figure(data=[trace_sales, trace_profit])

# Show figure
fig.show()
```

## 4. Create an interactive stacked bar chart showing the revenue by state (bars) for each of the products categories for California, New York and Texas.

```python
df_states = df[df['State'].isin(['California', 'New York', 'Texas'])]

grouped = df_states.groupby(['State', 'Category'])['Sales'].sum().unstack().reset_index()

categories = grouped.columns[1:]
colors = ['blue', 'orange', 'green', 'red']

data = []
for i, state in enumerate(['California', 'New York', 'Texas']):
    state_data = grouped[grouped['State'] == state]
    state_bar = go.Bar(x=state_data.columns[1:],
                       y=state_data.values.tolist()[0][1:],
                       name=state,
                       marker=dict(color=colors[i]))
    data.append(state_bar)

layout = go.Layout(title='Revenue by State for Each Product Category',
                   xaxis=dict(title='Product Category'),
                   yaxis=dict(title='Revenue'),
                   barmode='stack') 

fig = go.Figure(data=data, layout=layout)

fig.show()
```

## 5. Create an interactive line chart showing sale by day for California.

```python
df_california = df[df['State'] == 'California']

sales_by_day = df_california.groupby(df_california['Order Date'].dt.date)['Sales'].sum().reset_index()

trace = go.Scatter(x=sales_by_day['Order Date'],
                   y=sales_by_day['Sales'],
                   mode='lines',
                   name='Sales')

layout = go.Layout(title='Sales by Day in California',
                   xaxis=dict(title='Date'),
                   yaxis=dict(title='Sales'),
                   hovermode='closest')

fig = go.Figure(data=[trace], layout=layout)

fig.show()
```
