# Occupation

```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users 
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")[["age"]].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#Mi forma

male = users[users.loc[:,"gender"] == "M"]
male_ratio = male["occupation"].value_counts() / users["occupation"].value_counts()
male_ratio.sort_values(ascending=False)
```

```python
#Forma JL
proportions = users.groupby( "occupation")[["gender"]].value_counts(normalize=True).reset_index()
proportions.loc[proportions.loc[:, "gender"]=="M", :].sort_values("proportion", ascending=False)
```

```python
#Forma larga, "facil" que no hace falta saber mucho de Python
ratios = []
occus = []

for occu in users.loc[:,"occupation"].unique():
    subset = users.loc[users["occupation"] == occu, :]
    male = subset.loc[subset["gender"] == "M"].shape[0]
    female = subset.loc[subset["gender"] == "F"].shape[0]
    ratio = (male/(male + female)) * 100
    occus.append(occu)
    ratios.append(ratio)

pd.DataFrame({"occupation": occus, "ratio": ratios}).sort_values("ratio", ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")[["age"]].aggregate(['min', 'max'])
```

```python
#Otra manera
us = users.groupby("occupation")[["age"]].aggregate(['min', 'max'])
us.columns = ["Min age", "Max age"]
us
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"])["age"].mean().reset_index
```

### Step 8.  For each occupation present the percentage of women and men

```python
#Forma Nogue
us = users.groupby(["occupation", "gender"])[["gender"]].count()
# us.columns = ["Gender"]
# us = us.reset_index()
us1 = users.groupby("occupation")[["gender"]].count()
result = us/us1
result.columns = ["Ratio"]
result.reset_index()
```

```python
#Forma ChatGPT
occupation_gender = users.groupby('occupation')['gender']

gender_counts = occupation_gender.value_counts().unstack()

total_counts = gender_counts.sum(axis=1)

percentage_women = (gender_counts['F'] / total_counts) * 100
percentage_men = (gender_counts['M'] / total_counts) * 100

occupation_gender_percentage = pd.DataFrame({'Percentage of Women': percentage_women, 'Percentage of Men': percentage_men})

occupation_gender_percentage
```
